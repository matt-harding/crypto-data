```python
import pandas as pd
```

```python
df = pd.read_excel('Provider_Market_Data.xlsx')
```

```python
df
```

```python
coins = df.name.unique()
coins
```

```python
DataFrameDict = {elem : pd.DataFrame() for elem in coins}

for key in DataFrameDict.keys():
    DataFrameDict[key] = df[:][df.name == key]
```

```python
for key in ['Tether', 'Binance USD', 'Dai', 'USD Coin', 'Pax Dollar']:
    ax1 = DataFrameDict[key].plot.scatter(x='data_provider', y='price', title=key )
```

```python
for key in ['Tether', 'Binance USD', 'Dai', 'USD Coin', 'Pax Dollar']:
    ax1 = DataFrameDict[key].plot.scatter(x='data_provider', y='circulating_supply', title=key )
```

```python
for key in ['Tether', 'Binance USD', 'Dai', 'USD Coin', 'Pax Dollar']:
    ax1 = DataFrameDict[key].plot.scatter(x='data_provider', y='market_cap', title=key )
```
