```python
# https://pypi.org/project/coinmetrics-api-client/
from coinmetrics.api_client import CoinMetricsClient

API_KEY = 'INSERT_API_KEY'

# Community API:
client = CoinMetricsClient(API_KEY)
```

```python
# The price of an asset quoted in U.S. dollars using a framework to select high quality constituent markets and a methodology that is resistant manipulation.
metrics = ['ReferenceRate']
assets = ['usdt','usdc','busd','usdp','frax','dai','tusd','tusd']
```

```python
def fetch_data (assets, metrics):
    df = client.get_asset_metrics(
        assets = assets,
        metrics = metrics,
        start_time = '2022-05-01',
        end_time = '2022-05-31',
        frequency='1h'
    ).to_dataframe()
    
    return df
```

```python
import pandas as pd
```

```python
frames = []

for asset in assets:
    data = fetch_data([asset], metrics)
    data['Asset'] = asset
    frames.append(data)
    
result = pd.concat(frames)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
```

```python
plt.figure(figsize=(20,5))
for frame in frames:
    sns.lineplot(data = frame, x="time", y="ReferenceRate", label = frame['Asset'][0])
    
plt.show()
```

```python
abs_peg_diff = {}

for frame in frames:
    diff = 0
    for x in frame['ReferenceRate']:
        diff += abs(1 - x)
        
    abs_peg_diff[frame['Asset'][0]] = diff
    
abs_peg_diff
```
